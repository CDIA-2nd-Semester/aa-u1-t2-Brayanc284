**Ejercicio 1 – Producto punto vs. producto Hadamard**    
Genere dos matrices aleatorias A y B de tamaño 200 × 200:  
a) Calcule el producto Hadamard A ∗ B.  
b) Calcule el producto punto AB.  
c) Con %timeit compare el tiempo medio de ambas operaciones y discuta la diferencia teórica.  

```python
# Paso 1: Importamos las librerías necesarias
import numpy as np

# Paso 2: Generamos dos matrices aleatorias de tamaño 200x200
A = np.random.rand(20, 20)
B = np.random.rand(20, 20)

# Paso 3: Producto Hadamard (elemento a elemento)
hadamard = A * B  # Esto multiplica cada elemento con su correspondiente

# Paso 4: Producto punto (matricial)
dot = A @ B  # También se puede usar np.dot(A, B)

# Paso 5: Medimos el tiempo de ejecución con %timeit
# Solo funciona en Jupyter/Colab
print("Producto Hadamard:")
%timeit A * B

print("Producto punto:")
%timeit A @ B
```

**Ejercicio 4 – Derivada numérica por diferencias centradas**  
Sea f(x) = sin x e−x/4.  
Aproximar f′(x) en el intervalo [0, 10] usando diferencias centradas
de orden 2 con 500 puntos y comparar con la derivada analítica.


```python
# Paso 1: Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Paso 2: Definir la función f(x)
def f(x):
    return np.sin(x) * np.exp(-x / 4)

# Paso 3: Crear un vector de 500 puntos entre 0 y 10
x = np.linspace(0, 10, 500)
h = x[1] - x[0]  # Paso entre puntos (uniforme)

# Paso 4: Calcular la derivada numérica con diferencias centradas
# f'(x) ≈ (f(x+h) - f(x-h)) / (2h)
f_prime_num = (f(x[2:]) - f(x[:-2])) / (2 * h)

# Paso 5: Calcular la derivada analítica real
def f_prime_analitica(x):
    return np.cos(x) * np.exp(-x/4) - (1/4) * np.sin(x) * np.exp(-x/4)

# Paso 6: Ajustamos x para que coincidan los tamaños
x_mid = x[1:-1]
f_prime_exact = f_prime_analitica(x_mid)

# Paso 7: Graficar ambas derivadas
plt.figure(figsize=(10, 6))
plt.plot(x_mid, f_prime_num, label='Derivada Numérica', linestyle='--')
plt.plot(x_mid, f_prime_exact, label='Derivada Analítica', linestyle='-')
plt.title('Comparación de derivadas')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()
plt.grid(True)
plt.show()
```

**Ejercicio 7 – Estimación de parámetros de una normal**  
Simule 10 000 muestras de N(µ = 2,5, σ = 1,3). Estime µ y σ^2con fórmulas vectorizadas y
compare con np.mean y np.var.

```python
# Paso 1: Importamos NumPy
import numpy as np

# Paso 2: Fijamos la semilla para resultados reproducibles
np.random.seed(0)

# Paso 3: Simulamos 10,000 datos de una normal con media 2.5 y desviación 1.3
mu_real = 2.5
sigma_real = 1.3
datos = np.random.normal(loc=mu_real, scale=sigma_real, size=10000)

# Paso 4: Estimamos la media de forma manual (vectorizada)
media_manual = np.sum(datos) / len(datos)

# Paso 5: Estimamos la varianza de forma manual (con corrección n, no n-1)
varianza_manual = np.sum((datos - media_manual) ** 2) / len(datos)

# Paso 6: Estimamos con funciones de NumPy
media_numpy = np.mean(datos)
varianza_numpy = np.var(datos)

# Paso 7: Mostramos los resultados
print("Estimaciones manuales:")
print("Media:", media_manual)
print("Varianza:", varianza_manual)

print("\nEstimaciones con NumPy:")
print("Media:", media_numpy)
print("Varianza:", varianza_numpy)
```
